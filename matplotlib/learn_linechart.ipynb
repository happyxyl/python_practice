```python
# import library
import matplotlib.pyplot as plt
```

```python
# prepare data
x =  ["Mon", "Tues", "Wed", "Thur", "Fri", "Sat", "Sun"]
y = [20, 40, 35, 55, 42, 80, 50]
```

```python
# draw picture
plt.plot(x, y, "g", marker = 'D', markersize=5, label="Weekly Activity")
plt.show()
```

```python
# optimize picture
plt.plot(x, y, "g", marker = 'D', markersize=5, label="Weekly Activity")
plt.title("Active users")
plt.xlabel("Date")
plt.ylabel("Volume")
plt.legend()
for x1,y1 in zip(x, y) : plt.text(x1, y1, str(y1), ha='center', va='bottom', fontsize=15)
plt.show()
```
